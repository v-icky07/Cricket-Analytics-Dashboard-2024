```python
import pandas as pd
```

```python
# import os
# current_directory = os.getcwd()
# print(f"Current working directory: {current_directory}")
df_match = pd.read_csv("../RawCsv_2024/match_data.csv")
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename(columns={'Scorecard': 'Match_Id'}, inplace = True)
df_match.head()
```

**Create a match ids dictionary that maps team names to a unique match id. This will be useful later on to link with other tables**

```python
df_match.columns = df_match.columns.str.lower()
df_match.head()
```

```python
df_match.rename(columns={'team 1': 'team1','team 2': 'team2'}, inplace = True)
df_match.head()
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv('../t20_csv_files/dim_match_summary_2024.csv', index = False)
```
